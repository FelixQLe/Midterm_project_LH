#### **Task 4**: How taxi times changing during the day? Does higher traffic lead to bigger taxi times?

```python
import pandas as pd
import numpy as np
import datetime as dt
#from datetime import datetime
```

```python
data = pd.read_csv('data/Merged_Table_1.csv')
```

```python
data.head(5)
```

```python
"""
extracting useful columns
"""
taxi_df = data[['fl_date','taxi_out', 'taxi_in','origin_city_name','crs_dep_time','dest_city_name','crs_arr_time', 'Trip_count_PerDay']]
```

```python
#import TimeFormatted from function_helpers in folder data(change folder accordingly)
from data.function_helpers import TimeFormatted
```

```python
"""
extract hour from int format in 'crs_dep_time','crs_arr_time' columns
"""

taxi_df['crs_arr_time'] = taxi_df['crs_arr_time'].apply(TimeFormatted)
taxi_df['crs_dep_time'] = taxi_df['crs_dep_time'].apply(TimeFormatted)
taxi_df['crs_arr_time'] = pd.to_datetime(taxi_df['crs_arr_time'], format='%H:%M:%S').dt.minute
taxi_df['crs_dep_time'] = pd.to_datetime(taxi_df['crs_dep_time'], format='%H:%M:%S').dt.minute
```

```python
"""
As we can see higher traffic lead to bigger taxi out, taxi in times
"""
taxi_df.groupby(['origin_city_name','dest_city_name','crs_dep_time','crs_arr_time']).sum()
```
